```python
 # Filter for Zimbabwe and Afghanistan for years after 1952
countries_data = df[(df['country'].isin(['Zimbabwe', 'Afghanistan'])) & (df['year'] > 1952)]
```

```python
# Verify the data
print(countries_data[['country', 'year', 'lifeExpectancy']].head())
```

```python
# Pivot the data for easier plotting
pivot_df = countries_data.pivot(index='year', columns='country', values='lifeExpectancy')

# Plotting life expectancy trends for Zimbabwe and Afghanistan
pivot_df.plot(kind='line', marker='o')
plt.title('Life Expectancy in Zimbabwe and Afghanistan after 1952')
plt.xlabel('Year')
plt.ylabel('Life Expectancy (Age)')
plt.grid(True)
plt.show()
```

```python
zimbabwe_life_expectancy = countries_data[countries_data['country'] == 'Zimbabwe']['lifeExpectancy']
afghanistan_life_expectancy = countries_data[countries_data['country'] == 'Afghanistan']['lifeExpectancy']
```

```python
# Plot histograms
plt.hist(zimbabwe_life_expectancy, alpha=0.5, label='Zimbabwe', bins=10, edgecolor='black')
plt.hist(afghanistan_life_expectancy, alpha=0.5, label='Afghanistan', bins=10, edgecolor='black')

# Adding titles and labels
plt.title('Life Expectancy for Zimbabwe and Afghanistan after 1952')
plt.xlabel('Life Expectancy (Age)')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```
