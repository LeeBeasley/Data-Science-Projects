```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
diamonds = sns.load_dataset('diamonds')
```

```python
print(diamonds.head())  # Displays the first few rows of the dataset
print(diamonds.info())  # shows summary of the DataFrame
```

```python
# Statistical summary of the 'price' column
print(diamonds['price'].describe())
```

```python
print("Mode:", diamonds['price'].mode()[0])
```

```python
# histogram to view the distribution of prices
plt.figure(figsize=(10, 6))
sns.histplot(diamonds['price'], kde=True, color='blue')
plt.title('Distribution of Diamond Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#boxplot 
plt.figure(figsize=(10, 6))
sns.boxplot(x=diamonds['price'], color='green')
plt.title('Box Plot of Diamond Prices')
plt.xlabel('Price')
plt.show()
```

```python
#scatterplot show how the weight(carat) of the diamonds affects price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', data=diamonds)
plt.title('Price vs Carat')
plt.xlabel('Carat (Weight of the diamond)')
plt.ylabel('Price in USD')
plt.show()
```
